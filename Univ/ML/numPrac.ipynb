Numpy 이용해서 선형 방정식 풀이, 행렬식
10x+30y+2z=25
3x+y+z=-100
y+3z=20

```python
a=np.array([[10,3,2],[3,1,1],[0,1,3]])
b=np.array([30,15,20])
x=np.linalg.solve(a,b)
print(x)
```

Numpy를 활용한 Central Limit Theorem 실험

```python
np.random.seed(10)
np.random.rand(3)
```

```python
mu=10.0
sigma=10.0

_rand=np.random.normal(mu,sigma,size=10)
sns.histplot(_rand)
```

```python
_rand=np.random.normal(mu,sigma,size=50)
sns.histplot(_rand)
```

```python
_rand=np.random.normal(mu,sigma,size=100)
sns.histplot(_rand)
```

```python
_rand=np.random.normal(mu,sigma,size=200)
sns.histplot(_rand)
```

```python
_rand=np.random.normal(mu,sigma,size=10000*100)
sns.histplot(_rand)
```

```python
np.mean(_rand)
```

```python
np.median(_rand)
#평균과 중간값 중에 관측치가 적다면 평균에 대한 통계량을 얘기할 떄 일반적으로 중간값이 mean값보다 정확하다 
```

```python
plt.rc('font', family='Malgun Gothic')

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = dataset['pregnant']
y = dataset['plasma']
z = dataset['insulin']

ax.scatter(x, y, z)
ax.set_xlabel("임신 횟수")
ax.set_ylabel("공복 혈당 농도")
ax.set_zlabel("인슐린 수치")

plt.show()
```

```python
corr=dataset.corr()
sns.heatmap(corr,annot=True)
```

```python
mask=np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(10, 8))
    ax = sns.heatmap(corr, mask=mask, vmax=.3, annot=True, cmap='YlGnBu', square=True)
```

```python
X=dataset.iloc[:,:8]
X.head()

pg_plot=sns.PairGrid(X)
pg_plot.map_diag(sns.violinplot) # 특정 변수에 어떠한 값에 관측치가 많이 쏠려 있는지 표현 (distribution 대시 표현)
pg_plot.map_lower(sns.kdeplot) # kernel density estimation 으로 분포 추정, contour 그려봄
```
